Importamos las librerías necesarias

```python
import pandas as pd
import numpy as np
```

Extraemos el csv de origen

```python
inf = pd.read_csv('steam_games.csv')
inf
```

```python
inf.info()
```

Nos fijamos los valores únicos de la columna price.

```python
unique_prices = inf['price'].unique()
unique_prices
```

Nos fijamos los valores nulos de la misma columna

```python
filas_con_nulos = inf['price'].isnull().sum()
filas_con_nulos
```

Vemos cuales son los juegos que no tiene valores

```python
filas_con_nulos = inf[inf['price'].isnull()]
filas_con_nulos
```

Les damos el atributo "Free to play" a los juegos que no tienen precios

```python
inf['price'].fillna("Free to play", inplace=True)
```

Nos fijamos nuevamente los valores nulos para asegurarnos que se cambiaron

```python
filas_con_nulos = inf['price'].isnull().sum()
filas_con_nulos
```

Una vez hecha la transformacion disponibilizamos el csv para utilizarlo en lo codigos

```python
inf.to_csv('csv_final.csv', index = False)
```
